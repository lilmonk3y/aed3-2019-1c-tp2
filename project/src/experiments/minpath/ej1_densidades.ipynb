```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Se generaron 10 grafos de n vértices para n en [3, 20] saltando de 5 en 5
# y con probabilidad 0.75 de que un vértice dado exista.

# Luego se corre cada grafo 20 veces para evitar variaciones de tiempos del SO.

dfLow = pd.read_csv("./salidas/exp1/tiempos_1_salida_grafos_random_10_10_20_2_0.0.csv") # Creo un dataframe desde el csv
dfLow = dfLow.groupby("n", as_index=False).mean() # Promedio para cada n en milisegundos
print(dfLow)

dfMedium = pd.read_csv("./salidas/exp1/tiempos_2_salida_grafos_random_10_10_20_2_0.5.csv") # Creo un dataframe desde el csv
dfMedium = dfMedium.groupby("n", as_index=False).mean() # Promedio para cada n en milisegundos
print(dfMedium)

dfComplete = pd.read_csv("./salidas/exp1/tiempos_2_salida_grafos_random_10_10_20_2_0.5.csv") # Creo un dataframe desde el csv
dfComplete = dfComplete.groupby("n", as_index=False).mean() # Promedio para cada n en milisegundos
print(dfComplete)

```

```python
fig = plt.figure()
eje_cartesiano = sns.lineplot(x="n", y="bf", data=dfLow, color="blue");
eje_cartesiano = sns.lineplot(x="n", y="fw", data=dfLow, color="red");
eje_cartesiano = sns.lineplot(x="n", y="dp", data=dfLow, color="orange");
eje_cartesiano = sns.lineplot(x="n", y="df", data=dfLow, color="green");

eje_cartesiano.set_title("Densidad 0")
plt.ylabel("Tiempo (ms)");
plt.xlabel("Cantidad de vértices");
eje_cartesiano.legend(["Bellman-Ford", "Floyd-Warshall", "Dijkstra con cola de prioridad", "Dijkstra con cola FIFO"]);
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Se generaron 10 grafos de n vértices para n en [3, 20] saltando de 5 en 5
# y con probabilidad 0.75 de que un vértice dado exista.

# Luego se corre cada grafo 20 veces para evitar variaciones de tiempos del SO.

dfShort = pd.read_csv("./salidas/exp2/tiempos_4_salida_grafos_random_10_10_20_2_0.0_1_15.csv") # Creo un dataframe desde el csv
dfShort = dfShort.groupby("n", as_index=False).mean() # Promedio para cada n en milisegundos
print(dfShort)

dfLong = pd.read_csv("./salidas/exp2/tiempos_5_salida_grafos_random_10_10_20_2_0.0_16_60.csv") # Creo un dataframe desde el csv
dfLong = dfLong.groupby("n", as_index=False).mean() # Promedio para cada n en milisegundos
print(dfLong)


```

# Comparación

```python
fig = plt.figure()
eje_cartesiano = sns.lineplot(x="n", y="bf", data=dfShort, color="blue");
eje_cartesiano = sns.lineplot(x="n", y="fw", data=dfShort, color="red");
eje_cartesiano = sns.lineplot(x="n", y="dp", data=dfShort, color="orange");
eje_cartesiano = sns.lineplot(x="n", y="df", data=dfShort, color="green");

#eje_cartesiano.set_title("Distancias cortas")
plt.ylabel("Tiempo (ms)");
plt.xlabel("Cantidad de vertices");
eje_cartesiano.legend(["Bellman-Ford", "Floyd-Warshall", "Dijkstra con cola de prioridad", "Dijkstra con cola FIFO"]);
```
